## Part 2 — Cost of Living and Salary Analysis

In this section, I estimate my monthly cost of living for three locations where I might want to live after completing the MSBA program:

- San Diego, CA
- San Jose, CA
- San Francisco, CA

For each city, I estimate monthly expenses across major categories and then calculate:

1. **Minimum annual salary needed**
2. **Desired annual salary**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# assumptions
tax_rate = 0.25              # estimated effective tax rate
savings_rate_min = 0.10      # minimum savings goal
savings_rate_target = 0.20   # desired savings goal

cost_data = {
    "San Diego": {
        "Housing": 2400,
        "Food": 600,
        "Transportation": 250,
        "Utilities": 180,
        "Health/Insurance": 250,
        "Miscellaneous": 400
    },
    "San Jose": {
        "Housing": 3000,
        "Food": 650,
        "Transportation": 250,
        "Utilities": 200,
        "Health/Insurance": 250,
        "Miscellaneous": 450
    },
    "San Francisco": {
        "Housing": 3200,
        "Food": 700,
        "Transportation": 250,
        "Utilities": 220,
        "Health/Insurance": 250,
        "Miscellaneous": 500
    }
}
```

```python
cost_df = pd.DataFrame(cost_data).T

# total monthly and annual living costs
cost_df["Total Monthly Cost"] = cost_df.sum(axis=1)
cost_df["Annual Living Cost"] = cost_df["Total Monthly Cost"] * 12

# Salary
cost_df["Minimum Annual Salary Needed"] = cost_df["Annual Living Cost"] / ((1 - tax_rate) * (1 - savings_rate_min))
cost_df["Desired Annual Salary"] = cost_df["Annual Living Cost"] / ((1 - tax_rate) * (1 - savings_rate_target))

cost_df
```

### Interpretation of Results

The table above shows estimated monthly living expenses for three California locations I would consider after graduation: San Diego, San Jose, and San Francisco. Below are metrics that contribute to how location can affect salary expectations after graduation.

For each city, I calculated:

- **Total Monthly Cost**: the sum of all monthly living expenses
- **Annual Living Cost**: total monthly cost multiplied by 12
- **Minimum Annual Salary Needed**: the salary required to cover estimated living expenses while saving 10% of income after taxes
- **Desired Annual Salary**: the salary required to cover estimated living expenses while saving 20% of income after taxes

```python
summary_df = cost_df[["Total Monthly Cost", "Annual Living Cost",
                      "Minimum Annual Salary Needed", "Desired Annual Salary"]]

summary_df
```

```python
# Simple plot
summary_df["Desired Annual Salary"].plot(kind="bar", figsize=(8,5))

plt.title("Desired Annual Salary by Location")
plt.ylabel("Salary ($)")
plt.xlabel("Location")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### Brief Discussion

Based on my estimates, **San Diego** has the lowest cost of living among the three locations, while **San Francisco** has the highest. **San Jose** falls in between, but it is still significantly more expensive than San Diego.

Housing is the biggest driver of the difference across locations. Even though transportation, healthcare, and utilities are somewhat similar, rent increases total monthly expenses substantially in the Bay Area.

This analysis shows that salary expectations should depend on location. A salary that may feel comfortable in San Diego may not provide the same quality of life in San Jose or San Francisco. Because of that, it is important to think about both cost of living and salary opportunities together when evaluating post-graduation roles.
